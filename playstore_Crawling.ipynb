#### 배달의 민족

```python
from google_play_scraper import Sort, reviews_all
from datetime import datetime
import pandas as pd

lang = 'ko' # 언어: 한국어
cont = 'kr' # 국가: 한국

result = reviews_all ( # reviews_all -> 리뷰를 숫자 제한 없이 출력
    'com.sampleapp', # 'sampleapp' -> 배달의 민족 고유 id
    sleep_milliseconds = 0,
    lang=lang,
    country=cont,
    sort=Sort.NEWEST # 리뷰 최신순 정렬
)

start_date = datetime(2021, 1, 1)
end_date = datetime(2024, 9, 1)

# 필터링 된 데이터 담을 리스트
sampleapp_reviews = []

for item in result:
  review_date = item['at']

  if start_date <= review_date <= end_date:
    review = {
        'date': review_date.strftime("%Y.%m.%d"),
         #'userName' : item['userName'],
        'score' : item['score'],
        'content' : item['content']
    }
    sampleapp_reviews.append(review)

    # 날짜 제대로 필터 되는지 확인
    # print(review_date)

# 데이터프레임 형식으로 저장
sampleapp_reviews_df = pd.DataFrame(sampleapp_reviews)
print(len(sampleapp_reviews_df))
```

##### 파일 저장

```python
csv_filename = f'sampleapp_reviews_{(start_date).strftime("%Y.%m.%d")}_to_{(end_date).strftime("%Y.%m.%d")}.csv'
sampleapp_reviews_df.to_csv(csv_filename, index=False, encoding='utf-8-sig', escapechar='\\')
```
