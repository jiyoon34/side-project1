**리뷰 데이터 정규화**

```python
import re

# 정규화 함수 정의
def normalize_text(text):
    # 한국어, 숫자, 띄어쓰기만 남기고 나머지 문자는 제거
    return re.sub(r'[^가-힣0-9\s]', '', text)

# 'content' 열에 정규화 함수 적용하여 'content_re' 열 생성
coupang_reviews_df['content_re'] = coupang_reviews_df['content'].apply(normalize_text)
sampleapp_reviews_df['content_re'] = sampleapp_reviews_df['content'].apply(normalize_text)
```

```python
import pandas as pd
from konlpy.tag import Okt

# Okt 형태소 분석기 초기화
okt = Okt()

# 사용자 사전에 추가할 단어 목록 (분리되지 않도록 처리할 단어들)
user_dict = ['배달비','배송비','배달료','배송료','배송음식','배달음식','배송시간','배달시간','배달기사','배송기사','배달원','빠른배달','빠른배송','기사님','배달부','배달앱','자동차배달','배달현황','배송현황','배달위치','실시간','배달업체','배달계','로켓배송','배달도착','도착시간','예상도착시간','할인쿠폰','배달팁','배달주소','배달오류','배달지연',
             '결제방법','자동선택','요청사항','결제수단','결제옵션','주문내역','제휴카드','음식가격','첫주문','카드등록','진행상황','고객대응','가맹점','주문서','쿠팡이츠','쿠팡앱','쿠팡멤버십','와우클럽','쿠팡회원','쿠팡와우','쿠팡와우회원','쿠팡회비','쿠팡포인트','회원료','와우회원', '주문취소','쿠팡페이','쿠페이'
             '와우회원권','정기구독','쿠팡플레이','요기요','배민','배달의민족','타업체','타어플','다른앱','플랫폼','다른어플','유료회원','예정시간','만원쿠폰','쿠폰가격','뒷처리','할인행사','최소주문금액','최소금액','와우회원제','가성비','최소주문','댓글알바','매장방문','최소금액','친화적','합리적','주문량','주문량많은순',
             '재주문','화풀이','페널티','상담원','저녁시간','할인혜택','할인쿠폰', '본인명의','카드결제','일부매장','동일메뉴','무료배송','무료배달','무배','한집배달','무료화','패널티','구매자','소비자','네이버페이','별점','고객센터','상담원','전화연결','도둑놈','애용자','자영업자','주문누락','소액결제','카드결제',
             '선착순쿠폰','업체명','가게번호','주문순위','휴대폰결제','핸드폰결제','편의성','약속시간','예약주문','결제창','물품수정','사용요금','리프레쉬','법인카드','복잡','어수선','직관적','사용성','사용법','사용','사용자','실명제','카드결제','카드정보','요청사항','원탑','사장님','늦은시간','보관함','연계','중독성','유도리', '비마트', 'B마트','배달의민족','배달수준','한집배달','주문번호', '배민클럽','배달시간', '알뜰배달', '배민원', '포장료', '포장비', '배달주문','운영진', '배민페이', '삼성페이', '본인인증', '배민라이더스', '배달시스템']

# 형태소 분석을 수행하면서 사용자 사전 단어를 반영하는 함수 정의
def tokenize_content(text):
    morphs = okt.morphs(text)

    # 사용자 사전 단어가 분리되지 않도록 결합
    morphs_with_user_dict = []
    i = 0
    while i < len(morphs):
        word = morphs[i]

        # 단일 단어가 사용자 사전에 있으면 그대로 추가
        if word in user_dict:
            morphs_with_user_dict.append(word)
            i += 1  # 단일 단어 처리 시 1개만 스킵
        # 두 단어 혹은 세 단어 결합하여 사용자 사전에 있는지 확인
        elif i < len(morphs) - 2 and word + morphs[i + 1] + morphs[i + 2] in user_dict:
            morphs_with_user_dict.append(word + morphs[i + 1] + morphs[i + 2])
            i += 3  # 3개 단어를 결합했으므로 3개를 스킵
        elif i < len(morphs) - 1 and word + morphs[i + 1] in user_dict:
            morphs_with_user_dict.append(word + morphs[i + 1])
            i += 2  # 2개 단어를 결합했으므로 2개를 스킵
        else:
            morphs_with_user_dict.append(word)
            i += 1  # 단일 단어 처리 시 1개만 스킵

    return morphs_with_user_dict

# 쿠팡이츠 리뷰 데이터
data = {
    'date': coupang_reviews_df['date'].tolist(),
    'score': coupang_reviews_df['score'].tolist(),
    'content_re': coupang_reviews_df['content_re'].tolist()
}

# 데이터프레임 생성
okt_coupang_reviews_df = pd.DataFrame(data)

# 'content_re' 열의 모든 텍스트에 대해 형태소 분석 및 사용자 사전 반영
okt_coupang_reviews_df['okt_content'] = okt_coupang_reviews_df['content_re'].apply(tokenize_content)
```

```python
# 쿠팡이츠 리뷰 데이터
coupang_data = {
    'date': okt_coupang_reviews_df['date'].tolist(),
    'score': okt_coupang_reviews_df['score'].tolist(),
    'okt_content': okt_coupang_reviews_df['filtered_content'].tolist(),
    'filtered_content': okt_coupang_reviews_df['filtered_content'].tolist()
}

# 배민 리뷰 데이터
sampleapp_data = {
    'date': okt_sampleapp_reviews_df['date'].tolist(),
    'score': okt_sampleapp_reviews_df['score'].tolist(),
    'okt_content': okt_sampleapp_reviews_df['filtered_content'].tolist(),
    'filtered_content': okt_sampleapp_reviews_df['filtered_content'].tolist()
}


# 데이터프레임 생성
sw_coupang_reviews_df = pd.DataFrame(coupang_data)
sw_sampleapp_reviews_df = pd.DataFrame(sampleapp_data)


# 특정 단어들을 하나의 단어로 치환
def replace_words(words):
    return [
        '좋다' if word in ['좋아요', '좋다', '좋음', '좋아여', '좋습니다', '좋네요', '좋네여', '좋아', '좋은', '좋은데']
        else '좋겠다' if word in ['좋겠습니다', '좋겠어요', '좋을것같아요', '좋겠어']
        else word
        for word in words
    ]

def remove_stopwords(word_list, stop_words):
  filtered_words = []
  for word in word_list:
    if word not in stop_words:
      filtered_words.append(word)
  return filtered_words
```

```python
sw_coupang_reviews_df['filtered_content'] = sw_coupang_reviews_df['filtered_content'].apply(lambda x: remove_stopwords(x, coupang_stop_words))
#불용어처리 후에 발생되는 빈 열 삭제
sw_coupang_reviews_df = sw_coupang_reviews_df[sw_coupang_reviews_df['filtered_content'].apply(lambda x: len(x) > 0)]
sw_coupang_reviews_df
```
